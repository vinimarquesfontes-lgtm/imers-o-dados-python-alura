```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/guilhermeonrails/data-jobs/refs/heads/main/salaries.csv")
df.shape
```

```python
linhas, colunas = df.shape[0], df.shape[1]
print("linhas:", linhas)
print("colunas:", colunas)
```

```python
df.columns
```

```python
df.columns = ['ano', 'senioridade', 'contrato', 'cargo',
              'salario', 'moeda', 'usd', 'residencia',
              'remoto', 'empresa', 'tamanho_empresa']
display(df.head())
```

```python
df["senioridade"].value_counts()
```

```python
df["contrato"].value_counts()
```

```python
df["remoto"].value_counts()
```

```python
df["tamanho_empresa"].value_counts()
```

```python
senioridade = {
    'SE': 'Senior',
    'MI': 'Pleno',
    'EN': 'Jnior',
    'EX': 'Executivo'
}

df['senioridade'] = df['senioridade'].replace(senioridade)
display(df['senioridade'].value_counts())
```

```python
contrato= {
    'FT': 'Tempo Integral',
    'CT': 'Contrato',
    'PT': 'Meio Perodo',
    'FL': 'Freelancer'
}

df['contrato'] = df['contrato'].replace(contrato)
display(df['contrato'].value_counts())
```

```python
tamanho_empresa = {
    'M': 'Mdio',
    'L': 'Grande',
    'S': 'Pequeno'
}

df['tamanho_empresa'] = df['tamanho_empresa'].replace(tamanho_empresa)
display(df['tamanho_empresa'].value_counts())
```

```python
remoto = {
    0: 'Presencial',
    50: 'Hbrido',
    100: 'Remoto'
}

df['remoto'] = df['remoto'].replace(remoto)
display(df['remoto'].value_counts())
```

```python
import numpy as np

#criao de um dataframe para exemplo
df_salarios = pd.DataFrame({
  'nome': ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
   'salario': [4000, np.nan, 5000, np.nan, 100000]
})

#calcula a mdia salarial, substitui os nulos pela mdia e arredonda os valores
df_salarios['salario_media'] = df_salarios['salario'].fillna(df_salarios['salario'].mean().round(2))

'''Calcula a mediana salarial e substitui os nulos pela mediana'''
df_salarios['salario_mediana'] = df_salarios['salario'].fillna(df_salarios['salario'].median())

df_salarios
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 26, 27]
})

df_temperaturas['preenchido_ffill'] = df_temperaturas['Temperatura'].ffill()
df_temperaturas
```

```python
df_temperaturas = pd.DataFrame({
    'Dia': ['Segunda', 'Tera', 'Quarta', 'Quinta', 'Sexta'],
    'Temperatura': [30, np.nan, np.nan, 26, 27]
})

df_temperaturas['preenchido_bfill'] = df_temperaturas['Temperatura'].bfill()
df_temperaturas
```

```python
df_cidades = pd.DataFrame({
  'Nome' : ['Ana', 'Bruno', 'Carlos', 'Daniele', 'Val'],
  'cidade' : ['So Paulo', np.nan, 'Curitiba', np.nan, 'Belm']
  })
df_cidades['cidade_preenchida'] = df_cidades['cidade'].fillna('No informado')
df_cidades
```

```python
ordem_senioridade = ['Jnior', 'Pleno', 'Senior', 'Executivo']
plt.figure(figsize=(8,5))
sns.boxplot(x = 'senioridade', y ='usd', data = df_limpo, order = ordem_senioridade, palette = 'Set2', hue= "senioridade")
plt.title('Distribuio senioridade')
plt.xlabel('Salrio em U$D')
plt.show()
```

```python
import plotly.express as px
```

```python
media_salario_senioridade = df_limpo.groupby('senioridade')['usd'].mean().reset_index()
ordem_senioridade = ['Jnior', 'Pleno', 'Senior', 'Executivo']
fig = px.bar(media_salario_senioridade, x='senioridade', y='usd',
             title='Mdia Salarial por Senioridade',
             labels={'senioridade': 'Nvel de Senioridade', 'usd': 'Mdia Salarial Anual (U$D)'},
             category_orders={'senioridade': ordem_senioridade})
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values = 'quantidade',
             title = 'Proporo dos tipos de trabalho')
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values = 'quantidade',
             title = 'Proporo dos tipos de trabalho',
             hole= 0.5)
fig.show()
```

```python
remoto_contagem = df_limpo['remoto'].value_counts().reset_index()
remoto_contagem.columns = ['tipo_trabalho', 'quantidade']

fig = px.pie(remoto_contagem,
             names='tipo_trabalho',
             values = 'quantidade',
             title = 'Proporo dos tipos de trabalho',
             hole= 0.5)
fig.update_traces(textinfo = 'percent+label')
fig.show()
```

```python
df_ds = df_limpo[df_limpo['cargo'] == 'Data Scientist']

media_salario_pais = df_ds.groupby('empresa')['usd'].mean().reset_index()

fig = px.bar(media_salario_pais, x='empresa', y='usd',
             title='Salrio Mdio para Data Scientists por Pas',
             labels={'empresa': 'Pas', 'usd': 'Salrio Mdio Anual (U$D)'})
fig.show()
```
